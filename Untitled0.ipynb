```python
import pandas as pd
import numpy as np

df=pd.read_excel("/Covid_Dataset_new.xlsx",header=0)
df.head()
```

```python
print(df.shape)
df.info()
```

```python
df.describe(include="all")
```

```python
#feature selection
#missing data--->numeric=mean;;categoricl--mode();;;
#outlier treatment--->>incorrect data;

df.isnull().sum()
```

```python
column=['age_60_and_above','gender','test_indication','corona_result']

from sklearn import preprocessing
le=preprocessing.LabelEncoder()
for x in column:
  df[x]=le.fit_transform(df[x])
```

```python
df.head()
```

```python
x=df.values[:,:-1]#indipident varable array
y=df.values[:,-1]#dipident varable array
y=y.astype(int)
```

```python
#subset the array
#[rows,cols]
#x=[:0:-1]
#y=[:,-1]

from sklearn.model_selection import  train_test_split

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=10)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
#predecting using decision tree
from sklearn.tree import  DecisionTreeClassifier
#create model
model_DecisionTree=DecisionTreeClassifier(criterion="gini",random_state=10)
#train model
model_DecisionTree.fit(x_train,y_train)
#use model to pridict
y_pred=model_DecisionTree.predict(x_test)
print(y_pred)
```

```python
from sklearn.metrics import  confusion_matrix,accuracy_score
print(confusion_matrix(y_test,y_pred))
print(accuracy_score(y_test,y_pred))
```
